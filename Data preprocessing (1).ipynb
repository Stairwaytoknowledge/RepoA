```python
def create_top_partners(df, number_of_top_partners_each_category):
    
    df_top_partners = pd.DataFrame({'Nomenclature': [], 'ReporterISO3': [], 'ProductCode': [], 
                       'ReporterName': [], 'PartnerISO3': [], 'PartnerName': [],
                       'Year': [], 'TradeFlowName': [], 'TradeFlowCode': [], 'TradeValue in 1000 USD': []})
    
    for j in df["ReporterName"].unique():
        df_country = df[df["ReporterName"]==j]
        
        for i in range(1,8):
            df_by_country = df_country[df_country["ProductCode"]==i]
            df_by_country_sort = df_by_country.sort_values(by="TradeValue in 1000 USD", ascending=False)[:number_of_top_partners_each_category]
            df_top_partners = pd.concat([df_top_partners, df_by_country_sort], axis=0)
    
    return df_top_partners
```

```python
create_top_partners(df,10)
```

```python
df_top10_1996 = create_top_partners(df_1996,10)
df_top10_1997 = create_top_partners(df_1997,10)
df_top10_1998 = create_top_partners(df_1998,10)
df_top10_1999 = create_top_partners(df_1999,10)
```

```python
df_top10_1996_to_1999 = pd.concat([df_top10_1996,df_top10_1997,df_top10_1998,df_top10_1999], axis=0)
df_top10_1996_to_1999
```
